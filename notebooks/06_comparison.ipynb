# 06 — Model Comparison

**Purpose**: Compare LLM, XGBoost and JEPA-inspired model side by side.

All logic lives in `src/evaluation.py`. This notebook calls those functions and displays results inline.

**What we cover here**:
- Per-participant prediction table across all three models
- Metrics summary table
- ROC curves on the same chart
- MMSE trajectories for participants with actual CDR worsening
- Failure case analysis — where models disagree and why

```python
import sys
from pathlib import Path

project_root = Path().resolve().parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

%matplotlib inline
import matplotlib.pyplot as plt

from src.evaluation import (
    load_all_predictions,
    load_all_metrics,
    print_prediction_comparison_table,
    print_metrics_summary_table,
    print_failure_case_analysis,
    plot_roc_curves,
    plot_mmse_trajectories_for_positive_cases,
    save_comparison_table,
)

RESULTS_DIR  = project_root / "results"
FIGURES_DIR  = RESULTS_DIR / "figures"
RAW_DATA_PATH = project_root / "data" / "raw" / "oasis_longitudinal.csv"

print(f"Project root : {project_root}")
```

## 1. Load all predictions

```python
comparison_df = load_all_predictions(RESULTS_DIR)
comparison_df
```

## 2. Per-participant comparison table

Each row is one test participant. Columns show the actual outcome and each model's prediction.
Notes column flags the most interesting cases.

```python
print_prediction_comparison_table(comparison_df)
```

## 3. Metrics summary

```python
metrics_df = load_all_metrics(RESULTS_DIR)
print_metrics_summary_table(metrics_df)
```

## 4. ROC curves

```python
plot_roc_curves(comparison_df, FIGURES_DIR)
plt.show()
```

## 5. MMSE trajectories for worsening cases

For each participant who actually worsened, plot their full MMSE trajectory.
The dashed line marks the visit 2 prediction cutoff — everything to the right was held back.
Annotations show which models predicted correctly.

```python
plot_mmse_trajectories_for_positive_cases(comparison_df, RAW_DATA_PATH, FIGURES_DIR)
plt.show()
```

## 6. Failure case analysis

Where did models disagree? Where did a model catch a case others missed?
This is the centrepiece of the AMI Labs presentation narrative.

```python
print_failure_case_analysis(comparison_df)
```

## 7. Save comparison table
